```python
import os
import pprint
from spacy import displacy
from jupyter import RunJupyterDisplacy
from baseblock import Stopwatch

os.environ['ENABLE_SPACY_SPANNING'] = "false"
runner = RunJupyterDisplacy('nursing').process
```

```python
input_text = """

Urgent Care set to close tempoRARILY as nurses go on strike over FAIR paY

""".strip()

input_text
```

```python
sw = Stopwatch()
svcresult, displacy_results, options = runner(input_text)
pprint.pprint(svcresult)
print (str(sw))
```

```python
# options
```

```python
for item in displacy_results[:3]:
    html = displacy.render(item, style="ent", manual=True, options=options)
```
